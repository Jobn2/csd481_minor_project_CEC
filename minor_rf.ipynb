```python
import pandas as pd
import numpy as np
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
# Make sure Fake.csv and True.csv are uploaded in Colab
fake = pd.read_csv("Fake.csv", on_bad_lines='skip', encoding='latin-1')
real = pd.read_csv("True.csv", on_bad_lines='skip', encoding='latin-1')

fake['label'] = 0
real['label'] = 1

df = pd.concat([fake, real]).reset_index(drop=True)
print("Dataset loaded. Shape:", df.shape)
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # remove punctuation/numbers/special chars
    tokens = nltk.word_tokenize(text)  # tokenize words
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]  # remove stopwords + lemmatize
    return " ".join(tokens)

df["clean_text"] = df["text"].apply(preprocess_text)
print("✅ Preprocessing done")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["clean_text"], df["label"], test_size=0.2, random_state=42
)
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)
rf_model.fit(X_train_vec, y_train)
print("✅ Random Forest training complete")
```

```python
y_pred = rf_model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
